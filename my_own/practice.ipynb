```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import qutip

import pulser
from pulser import Pulse, Sequence, Register
from pulser_simulation import Simulation
from pulser.devices import MockDevice, Chadoq2
from pulser.waveforms import BlackmanWaveform
```

```python
# Times are in ns
t_pulse = 26
delta_field = 2.0

#the parameters here are 1) the Omega values; 2) the phase shifts; 3) the phase. 
X_pulse = Pulse.ConstantDetuning(BlackmanWaveform(t_pulse, np.pi / 2.0), delta_field, 0)
Y_pulse = Pulse.ConstantDetuning(
    BlackmanWaveform(t_pulse, np.pi / 2.0), delta_field, -np.pi / 2
)
#there are different by the constant phase they each have!
```

```python
coords = np.array([[0, 0], [10, 0]])
qubits = dict(enumerate(coords))
reg = Register(qubits)

seq = Sequence(reg, MockDevice)
seq.declare_channel("MW", "mw_global")
seq.set_magnetic_field(0.0, 0.0, 1.0)
seq.add(Y_pulse, 'MW')
seq.draw()
```
